# Password Picker using Python

## What is a Strong and Simple Password?

## Generate a Strong Password using the following Reference Flowchart.

```python
#Step1 - we will store lists with random adjectives, nouns and punctuations. 
#Step2 - we will generate random number between 0 and 1000 using random.randint 
#Step3 - we will combine the random adj, nouns,numbers and punctuations to create a fairly strong password.
#Step4 - display the password as a string 
```

```python
#list of English Adjectives 

adj = ['tired','quick','charming','gentle','sharp','interesting','strange','engaging','lazy',
      'kiddish','fast','dangerous','flawless','pretty','beautiful','strange','huge','holycat']
```

```python
#list of English Nouns 

nouns = ['cow','cat','france','mississippi','company','food','jelly','teddy','cricket',
      'bike','pastry','school','lion','sandwich','horse','raincoat','gun']
```

```python
#list of usable English Punctuations for legible password

punct = ['@','.','?','#','!',':','_']

#alternatively, we can also generate punctuations using string.punctuation after importing string library.
#import string 
#string.punctuation
```

```python
#importing random for random passoword picking

import random

def pass_pick():
    adjec = random.choice(adj)
    noun = random.choice(nouns)
    digit = random.randint(0,1000)
    symb = random.choice(punct)
    
    #concatenating the elements of the final password as a string
    password = adjec + noun + str(digit) + symb
    
    print("Initiating the Password Picker Program!\n")
    print("Hurray! Your password is {}".format(password))
    
pass_pick()
```

```python
#calling the password picker again
pass_pick()
```
